```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## Data Collection and Processing
```

```python
# Loading the csv data to a pandas 
patient_data = pd.read_csv('heart_disease.csv')
```

```python
# First 5 rows of the dataset
patient_data.head()
```

```python
# no. of rows and columns in the datasets
patient_data.shape
```

```python
# Getting some info about the data
patient_data.info()
```

```python
# Checking for missing values
patient_data.isnull()
```

```python
# Statistical measures about the data
patient_data.describe()
```

```python
## Features - Train model(How he looks like)(Characteristics, Attributes)
## Target - Prediction
# checking the distribution of target variable to see how many have heart diease and how many do not
patient_data['target'].value_counts()
```

```python
## Splitting the Features and Target
# Feature, x
x = patient_data.drop(columns = 'target', axis = 1)

# Target, y
y = patient_data['target']
```

```python
print(y)
```

```python
print(x)
```

```python
## Splitting the Data into Training data and Test data
# Stratify = y to make the distribution of the target values proportional
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
# Check the number of tests and training data we have gotten
# x = original dataset with the feau
print(x.shape, x_train.shape, x_test.shape)
```

print(y.shape, y_train.shape, y_test.shape)

## Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train.values, y_train.values )
```

## Model Evaluation

```python
x_train_prediction = model.predict(x_train.values)
training_data_accuracy = accuracy_score(x_train_prediction, y_train.values)
```

```python
print(x_train_prediction)
```

```python
print(f'Accuracy on Training data: {training_data_accuracy}')
```
